```python
def my_sin(x):
    return np.sin(x)

def my_cos(x):
    return np.cos(x)

x = np.linspace(-np.pi, np.pi)
y_sin = my_sin(x)
y_cos = my_cos(x)

plt.plot(x, y_sin, label='sin')
plt.plot(x, y_cos, label='cos')
plt.legend()

plt.xlabel('x', size=14)
plt.ylabel('y', size=14)
plt.grid()
plt.show()
```

```python
def my_tan(x):
    return np.tan(x)


x = np.linspace(-1.3, 1.3)
y_tan = my_tan(x)


plt.plot(x, y_tan, label='sin')
plt.legend()

plt.xlabel('x', size=14)
plt.ylabel('y', size=14)
plt.grid()
plt.show()
```

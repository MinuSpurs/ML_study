```python
import numpy as np
import matplotlib.pyplot as plt

x = []
y = []
total = 0
num_5 = 0   # 5가 나온 횟수
n = 5000    # 주사위를 던진 횟수

for i in range(n):
    if np.random.randint(6)+1 == 5:
        num_5 += 1

    total += 1
    x.append(i)
    y.append(num_5/total)

plt.plot(x, y)
plt.plot(x, [1/6]*n, linestyle='dashed')    # 1/6을 점선으로 표시

plt.xlabel('x', size=14)
plt.ylabel('y', size=14)
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = []
y = []
total = 0
num_front = 0   # 동전 앞면이 위가 된 횟수
n = 5000    # 동전을 던진 횟수

for i in range(n):
    if np.random.randint(2) == 0:
        num_front += 1

    total += 1
    x.append(i)
    y.append(num_front/total)

plt.plot(x, y)
plt.plot(x, [1/2]*n, linestyle='dashed')    # 1/2을 점선으로 표시

plt.xlabel('x', size=14)
plt.ylabel('y', size=14)
plt.grid()
plt.show()
```

```python
# 기대값 구현

p = np.array([0.8, 0.15, 0.05]) # 확률
x = np.array([100, 500, 1000])  # 값

print(np.sum(p*x))

# 기대값 = 평균 (같은 의미로 사용됨)
```

```python
# 행렬을 다루기 위한 모듈을 불러온다.
import numpy as np
# 데이터를 읽어오기 위한 pandas 모듈을 불러온다.
import pandas as pd 

np.random.seed(1)
p = 2
true_beta = np.array([[1], [0.5]])
n = 1000
x = np.random.normal(loc=0, scale=0.2, size=(n,1)) # x라는 설명변수를 정규 분포로부터 1000개를 뽑았음
x = np.hstack((np.ones((n,1)), x)) #상수항과 회귀항을 같이 분석하기 위해서 1이라는 벡터와 우리가 생성한 x벡터를 합치는 곳
x[:5, :]
```

```python
# 모수값을 계산하고 포아송 분포를 이용해 y를 생성
parm = np.exp(x @ true_beta)
parm[:5, :]
```

```python
y = np.random.poisson(parm)
y[:5, :]
```

```python
beta = np.array([.5, .5]).reshape((p, 1)) #beta값은 0.5, 0.5
#1차 미분
parm = np.exp(x @ beta)
grad = -np.mean(y*x - parm*x, axis=0).reshape((p,1)) #gradiend descent에서 사용할 변수
grad
```

```python
learning_rate = 0.7
#initial beta
beta = np.zeros((p, 1)) #초기 솔루션은 0으로부터 시작
for i in range(500):
    #모수추정값, 1차 미분 계산
    parm = np.exp(x @ beta)
    grad = -np.mean(y*x - parm*x, axis=0).reshape((p,1))
    
    #beta update
    beta_new = beta - learning_rate*grad
    
    #stopping rule, 해가 수렴할때 까지 반복
    if np.sum(np.abs(beta_new - beta)) < 1e-8:
        beta = beta_new
        print('Iteration {} beta: '.format(i+1))
        print(beta, '\n')
        break
    else:
        beta = beta_new
        print('Iteration {} beta: '.format(i+1))
        print(beta, '\n')
```

```python
beta = np.array([.5, .5]).reshape((p,1))
# 2차 미분
parm = np.exp(x @ beta)
D = np.diag(np.squeeze(parm))
D
```

```python
H = x.T @ D @ x/n
H
```

```python
# initial beta
beta = np.zeros((p, 1))
for i in range(500):
    # 모수추정값, 1차와 2차 미분 계산
    parm = np.exp(x @ beta)
    grad = -np.mean(y*x - parm*x, axis=0).reshape((p,1))
    D = np.diag(np.squeeze(parm))
    H = x.T @ D @ x/n
    
    # beta update
    beta_new = beta - np.linalg.inv(H) @ grad #역행렬을 구하는 줄이 추가되었음(해를 갱신할 때 H의 역행렬이 필요함)
    
    # stopping rule
    if np.sum(np.abs(beta_new - beta)) < 1e-8:
        beta = beta_new
        print('Iteration {} beta:'.format(i+1))
        print(beta, '\n')
        break
    else:
        beta = beta_new
        print('Iteration {} beta:'.format(i+1))
        print(beta, '\n')
```

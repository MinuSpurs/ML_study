```python
text = '기계학습 방법론 을 이해 하여 인공 지능 을 만들어 봅시다 .'

text_tokenized = text.split(' ')

vocab = {'.': 0,
         '기계학습': 1,
         '만들어': 2,
         '방법론': 3,
         '봅시다': 4,
         '을': 5,
         '이해': 6,
         '인공': 7,
         '지능': 8,
         '하여': 9}

text_sequence = [vocab.get(x) for x in text_tokenized]
text_sequence
```

```python
vocab_size = len(vocab)
vocab_size
```

```python
import tensorflow as tf

one_hot = tf.keras.utils.to_categorical(text_sequence, 
                                        num_classes=vocab_size, 
                                        dtype='int')
one_hot
```
